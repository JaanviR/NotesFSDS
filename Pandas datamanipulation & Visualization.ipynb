```python
np.random.seed(20)
arr_3 = np.random.randn(10,4) # creates 10*4 matrix with random standard normalization values
pd.DataFrame(arr_3)#converts the array or matrix into dataframe
df = pd.DataFrame(arr_3,index = pd.date_range('2021-07-19',periods = 10),columns =['a','b','c','d']) #assigning our own index and columns names
```

```python
df
```

```python
df["new_col"] = df['c']+df['d'] # creating a new column with values as sum of 'c' column and 'd' column
```

```python
df
```

```python
df["newcol2"]=df['d'].apply(lambda x: x**2) # creates a new column with square values of column 'd'
```

```python
df
```

```python
df["newcol3"]=df['d'].apply(lambda x: np.log(x)) # creates new column with logarthmic values of column 'd'
```

```python
df
```

```python
def test(x):
    if(x<0):
        return x
    else:
        return np.log(x)
    
df["newcol4"]=df['d'].apply(test) # creates new column with logarthmic values of column 'd' but here we will apply a function with some condition saying that if the 'd' values are less than 0 then keep the 'd' value as it is else return the logarithmic value of 'd'
```

```python
df
```

```python
#create a new column with values based on a condition(if 'a' value is -ve return 1 else return 0)
def check(x):
    if (x<0):
        return 1
    else:
        return 0

df["newcol5"]=df['a'].apply(check)
```

```python
df
```

```python
df.sort_values(by='b') # sort the entire dataframe based on column 'b'
```

```python
df["newcol6"]=df.index # creates a new column with values as "index" values
```

```python
df
```

```python
# splitting the date into 3 columns as day, month and year
df["day"]=df.index.day
df["month"]=df.index.month
df["year"]=df.index.year
```

```python
df
```

```python
df.groupby(["day","year"]).mean() # grouping the dataset based on 'day' and 'year' columns and diplaying mean values of each column
```

```python
df["weekday"]=df.index.day_name() # create a new column with day name
```

```python
df
```

```python
df.groupby("weekday").count() # group the entire dataset based on 'weekday' column and count how many values are present for each column on that particular day
```

```python
df2.plot(kind="scatter", x='sepal_length', y='petal_width')
#plotting a scatter graph, here we have to mention between what columns you want to plot the scatter graph
#x value is what should be on x-axis, y value is what should be on y-axis
#for eg: we plotted the below graph between the columns 'sepal_length & petal_width'
# Scatter plot is used to find out the relationship between any two or three columns
#the below graph is showing the relationship between 2 columns(sepal_length and petal_width)
#from the below graph we can understand that as sepal_length increases petal_width also increases
#similarly using scatter plots we can find other things like outliers, we can find clusters/groups etc.,
```

```python
#finding the relationship between 3 columns or finding the effect on one column based on another 2 columns
df2.plot(kind="scatter", x='sepal_length', y='petal_width',c="petal_length")
#in the below graph the third column(petal_length) is taken using color representation
# for eg: from the below graph we can conclude that for the sepal_lengths between 4.0 to 6.0 
#and 
#for petal_widths between 0.0 to approximately 0.6, the dot colors are light..because if you observe the petal_lenght bar
#the petal_lengths in the range between 1 and 3 have light colors..so from this we can say that 
#for the sepal_lengths between 4.0 to 6.0 and for petal_widths between 0.0 to approximately 0.6; "petal-lengths will be in the range of 1 and 3"
# in the same way we can also say that when
#sepal_length - 7.5 to 8 & petal_width - 2.0 to 2.3, in that particular region the petal_lenth is gonna be above 5 as the dots are very bright
```

```python
df2.plot(kind="scatter", x='sepal_length', y='petal_width',c="petal_length", label="jaanvi",s=80)
# label-->giving name to dots
# s-->size of dots
```

```python
# another kind of plot known as hexbin plot, here the dots are represented in the form of hexagons
df2.plot(kind="hexbin", x='sepal_length', y='sepal_width', label="jaanvi",gridsize=10)
```
